# Hyperopt-Sklearn on Convex Images

The ["Convex Images" data set](http://www.iro.umontreal.ca/~lisa/twiki/bin/view.cgi/Public/MnistVariations) was introduced to evaluate the capacity of deep neural networks to disentangle many factors of variation.

Using Hyperopt-sklearn, it is possible to obtain state-of-the-art scores on this data set using standard (non-deep) classification algorithms. The scores are better than the baselines reported in e.g. [Larochelle et al. 2007]() because we allow for PCA pre-processing, and because hyperparameter optimization is a difficult thing to do. It is also possible to improve the accuracy of deep networks by allowing for different preprocessing options and by applying hyperparameter optimization techniques.

The baselines presented here are using relatively fast algorithms. There are 10 000 training examples, and we can get good results even by restricting ourselves to models that train in less than 5 minutes. Many configurations require much less time. This notebook demonstrates optimization using 300 evaluation points. The entire search process takes approximately three hours to run, and usually finds a few configurations with around approx 87% accuracy.

```python
# -- RETRIEVE DATA VIA SKDATA (github.com/jaberg/skdata)
import skdata.larochelle_etal_2007.view

class DataCaptureMock(object):
    def best_model(self, train, valid):
        self.train = train
        self.valid = valid

    def loss(self, model, task, **kwargs):
        if task not in (self.train, self.valid):
            self.test = task
# store train, valid, test tasks into data_capture
data_capture = DataCaptureMock()
skdata.larochelle_etal_2007.view.ConvexVectorXV().protocol(data_capture)
```

```python
import hpsklearn
import hyperopt.tpe
estimator = hpsklearn.HyperoptEstimator(
    #preprocessing=simple_small_image_preprocessing('pp'),
    #classifier=hpc.any_classifier('classif'),
    max_evals=300,
    verbose=1,
    algo=hyperopt.tpe.suggest,
    trial_timeout=60.0 * 5, # -- seconds
    )
```

```python
# Demo version of estimator.fit()
import hpsklearn.demo_support

fit_iterator = estimator.fit_iter(data_capture.train.x, data_capture.train.y)
fit_iterator.next()
plot_helper = hpsklearn.demo_support.PlotHelper(estimator)
while len(estimator.trials.trials) < estimator.max_evals:
    fit_iterator.send(1) # -- try one more model
    plot_helper.post_iter()
plot_helper.post_loop()

# -- Model selection was done on a subset of the training data.
# -- Now that we've picked a model, train on all training data.
estimator.retrain_best_model_on_full_data(data_capture.train.x, data_capture.train.y)
```
